### 0. Import util libraries


```python
import pandas as pd
from matplotlib import rc

from inversion_sst_gp import plot_helper

rc("font", family="serif", serif=["Computer Modern"])
rc("text", usetex=True)
rc("text.latex", preamble=r"\usepackage{amsmath}")
```

### 1. Fully observed noiseless experiment - time series


```python
# Load scoring data
df_time_1h_gp_obs_t = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_1h_gp_obs_t.csv"
)
df_time_1h_gp_num_t = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_1h_gp_num_t.csv"
)
df_time_1h_gp_num_t1 = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_1h_gp_num_t1.csv"
)
df_time_1h_gos_t = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_1h_gos_t.csv"
)

df_time_24h_gp_obs_t = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_24h_gp_obs_t.csv"
)
df_time_24h_gp_num_t = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_24h_gp_num_t.csv"
)
df_time_24h_gp_num_t1 = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_24h_gp_num_t1.csv"
)
df_time_24h_gos_t = pd.read_csv(
    "2_covariance_parameter_estimation/outputs/time_24h_gos_t.csv"
)
```

```python
fig, ax = plot_helper.plot_time_metrics(
    df_time_1h_gp_obs_t,
    df_time_1h_gp_num_t,
    df_time_1h_gp_num_t1,
    df_time_1h_gos_t,
    df_time_24h_gp_obs_t,
    df_time_24h_gp_num_t,
    df_time_24h_gp_num_t1,
    df_time_24h_gos_t,
    return_fig=True,
)
fig.savefig(
    "3_observing_system_simulation_experiment/figures/osse_metrics_time_48h.png",
    bbox_inches="tight",
    dpi=300,
)
```
