```python
import xarray as xr
import numpy as np
import os
```

```python
# load geostrophic currents
LON_LIMITS = (115, 118)
LAT_LIMITS = (-15.5, -12.5)
path_geo = 'non_proc_data/altimetry'
PROCESSED_DIR = "1_preproc_data/proc_data"
geo_files = os.listdir(path_geo)
geo_files = [f for f in geo_files if f.endswith('.nc')]
geo_paths = [os.path.join(path_geo, f) for f in geo_files]
dsgeo_o = xr.open_mfdataset(geo_paths)


ds_geo_c = dsgeo_o.sel(longitude = slice(LON_LIMITS[0],LON_LIMITS[1]), latitude = slice(LAT_LIMITS[0],LAT_LIMITS[1]))
ds_geo = xr.Dataset(
    {
        'ugos': (['time', 'lat', 'lon'], ds_geo_c.ugos.values),
        'vgos': (['time', 'lat', 'lon'], ds_geo_c.vgos.values),
    },
    coords={
        "time": ds_geo_c.time.values,
        "lat": ds_geo_c.latitude.values,
        "lon": ds_geo_c.longitude.values
    }
)
ds_geo
ds_geo.to_netcdf(f'{PROCESSED_DIR}/altimeter_currents.nc')
```
